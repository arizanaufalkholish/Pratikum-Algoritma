```python
test_data = ['07:31']

for jam in test_data:
    hasil = jam
    jam_menit = int(hasil.replace(':', ''))

    if jam_menit > 830:
        kategori = "Absen tidak valid"
    elif jam_menit >= 731:
        kategori = "Terlambat"
    elif jam_menit >= 730:
        kategori = "Tepat Waktu"
    elif jam_menit >= 630:
        kategori = "Tepat Waktu"
    elif jam_menit >= 628:
        kategori = "Absen tidak valid"
    else:
        kategori = "Absen tidak valid"

    print(f"Jam: {jam} → {kategori}")
```

### Menggunakan data `laptop_price.csv` secara terpisah

```python
import pandas as pd
import numpy as np
try:
    file_path_laptop = '/content/laptop_price.csv'

    df_laptops = pd.read_csv(file_path_laptop, encoding='latin-1')
    print(f"File '{file_path_laptop}' berhasil dimuat ke DataFrame 'df_laptops'.")

    print("\n5 baris pertama dari df_laptops:")
    display(df_laptops.head())

    if 'TypeName' in df_laptops.columns:
        ultrabook_count = df_laptops[df_laptops['TypeName'] == 'Ultrabook'].shape[0]
        print(f"\nJumlah laptop dengan kategori 'Ultrabook': {ultrabook_count}")
    else:
        print("\nKolom 'TypeName' tidak ditemukan di df_laptops. Pastikan dataset memiliki kolom ini.")

except FileNotFoundError:
    print(f"Error: File '{file_path_laptop}' tidak ditemukan. Pastikan Anda sudah mengunggah file dengan nama yang benar dan path yang tepat.")
except Exception as e:
    print(f"Terjadi kesalahan saat memuat atau memproses file laptop: {e}")
```

```python
import numpy as np

if 'df_laptops' in locals() and 'Price_euros' in df_laptops.columns:
    df_laptops['price_category'] = np.where(df_laptops['Price_euros'] > 1000, 'Mahal', 'Tidak Mahal')
    print("Kolom 'price_category' berhasil dibuat.")
    display(df_laptops[['Price_euros', 'price_category']].head())
else:
    print("DataFrame 'df_laptops' atau kolom 'Price_euros' tidak ditemukan. Pastikan data laptop sudah dimuat dengan benar.")
```

```python
if 'df_laptops' in locals() and 'price_category' in df_laptops.columns:
    mahal_count = df_laptops[df_laptops['price_category'] == 'Mahal'].shape[0]
    print(f"Jumlah laptop dengan kategori harga 'Mahal': {mahal_count}")
else:
    print("DataFrame 'df_laptops' atau kolom 'price_category' tidak ditemukan. Pastikan data laptop sudah dimuat dan kategori harga telah dibuat.")
```

```python
OUNCES_PER_METRIC_TON = 35273.92

berat_kotak = float(input("Masukkan berat satu kotak cereal (dalam ounces): "))

berat_metric_ton = berat_kotak / OUNCES_PER_METRIC_TON


jumlah_kotak = OUNCES_PER_METRIC_TON / berat_kotak

print(f"\nBerat satu kotak cereal: {berat_kotak} ounces")
print(f"Berat dalam metric ton: {berat_metric_ton:.6f} metric tons")
print(f"Jumlah kotak untuk 1 metric ton: {jumlah_kotak:.2f} kotak")
```

```python
def hitung_kalori(met, berat_kg, durasi_menit):
    """
    Menghitung total kalori yang terbakar berdasarkan MET, berat, dan durasi.

    Parameters:
    met (float): Nilai MET untuk aktivitas
    berat_kg (float): Berat badan dalam kilogram
    durasi_menit (float): Durasi aktivitas dalam menit

    Returns:
    float: Total kalori yang terbakar
    """
    kalori_per_menit = 0.0175 * met * berat_kg
    total_kalori = kalori_per_menit * durasi_menit
    return total_kalori


def pounds_ke_kg(pounds):
    """Mengkonversi pounds ke kilogram"""
    return pounds / 2.2


# Contoh penggunaan program
def main():
    print("=" * 50)
    print("KALKULATOR KALORI BERDASARKAN MET")
    print("=" * 50)

    # Daftar aktivitas dan MET-nya
    aktivitas_met = {
        "1": ("Running 6 MPH", 10),
        "2": ("Basketball", 8),
        "3": ("Sleeping", 1)
    }

    # Tampilkan menu aktivitas
    print("\nPilih aktivitas:")
    for key, (nama, met) in aktivitas_met.items():
        print(f"{key}. {nama} ({met} METS)")

    # Input pilihan aktivitas
    pilihan = input("\nMasukkan nomor aktivitas (1-3): ")

    if pilihan not in aktivitas_met:
        print("Pilihan tidak valid!")
        return

    nama_aktivitas, met = aktivitas_met[pilihan]

    # Input berat badan
    satuan_berat = input("Berat badan dalam Kg atau Pounds? (kg/lb): ").lower()
    berat_input = float(input(f"Masukkan berat badan: "))

    if satuan_berat == "lb" or satuan_berat == "pounds":
        berat_kg = pounds_ke_kg(berat_input)
        print(f"Berat: {berat_input} pounds = {berat_kg:.2f} kg")
    else:
        berat_kg = berat_input

    # Input durasi
    durasi = float(input("Masukkan durasi aktivitas (menit): "))

    # Hitung kalori
    total_kalori = hitung_kalori(met, berat_kg, durasi)

    # Tampilkan hasil
    print("\n" + "=" * 50)
    print("HASIL PERHITUNGAN")
    print("=" * 50)
    print(f"Aktivitas      : {nama_aktivitas}")
    print(f"MET            : {met}")
    print(f"Berat Badan    : {berat_kg:.2f} kg")
    print(f"Durasi         : {durasi} menit")
    print(f"Total Kalori   : {total_kalori:.2f} kalori")
    print("=" * 50)


# Jalankan program
if __name__ == "__main__":
    main()
```
